# Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## I. Other algorithms

```python
path = '../data/'
df = pd.read_csv(path + 'results.csv')
```

```python
df['algorithm'] = df.algorithm.str[:] + ' - ' + df.type.str[:]
df.head()
```

## Croatia

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        if point['val'] == 'Fast Greedy - weighted':
            ax.text(point['x'], point['y']-0.03, str(point['val']))
        elif point['val'] == 'Walktrap - unweighted':
            ax.text(point['x'], point['y']-0.015, str(point['val']))
        elif point['val'] == 'Label propagation - weighted':
            ax.text(point['x']-2000, point['y']+0.015, str(point['val']))
        else:
            ax.text(point['x'], point['y']+0.015, str(point['val']))
        
```

```python
df_CR = df[(df.dataset == 'Deezer - Croatia')]
df_CR = df_CR.drop(22)
plt.figure(figsize=(13,7))
label_point(df_CR.nb_clusters, df_CR.modularity, df_CR.algorithm, plt.gca())
sns.set()
sns.scatterplot(x='nb_clusters', y='modularity', size='time',
                hue='algorithm', sizes=(200, 1000), legend='brief',
                data=df_CR, palette = 'bright')
plt.xscale('log')
plt.legend([])
plt.show()
```

## II. GEMSEC

```python
path = '../data/'
df2 = pd.read_csv(path + 'results_GEMSEC.csv')
df2.columns = ['algorithm', 'DW', 'nb_clusters', 'dataset', 'time', 'modularity']
```

```python
df2['algorithm'] = df2.algorithm.str[:] + ' + DW' + df2.DW.astype('str')
df2['time'] = df2.time * 60
df2.head()
```

## Croatia

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        if point['val'] == 'Fast Greedy - weighted':
            ax.text(point['x'], point['y']-0.03, str(point['val']))
        elif point['val'] == 'Walktrap - unweighted':
            ax.text(point['x'], point['y']-0.015, str(point['val']))
        elif point['val'] == 'Label propagation - weighted':
            ax.text(point['x']-2000, point['y']+0.015, str(point['val']))
        else:
            ax.text(point['x'], point['y']+0.015, str(point['val']))
```

```python
df_CR = df[(df.dataset == 'Deezer - Croatia')]
df_CR = df_CR.drop(22)
df_CR2 = df2[(df2.dataset == 'Deezer - Croatia')]
df_CR = pd.concat([df_CR, df_CR2], axis=0, sort=True)
plt.figure(figsize=(13,7))
label_point(df_CR.nb_clusters, df_CR.modularity, df_CR.algorithm, plt.gca())
sns.set()
sns.scatterplot(x='nb_clusters', y='modularity', size='time',
                hue='algorithm', sizes=(200, 1000), legend='brief',
                data=df_CR, palette = 'bright')
plt.xscale('log')
plt.legend([])
plt.show()
```

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    boolean = 1
    for i, point in a.iterrows():
        if (point['val'] == 'GEMSEC + Reg. + DW2'):
            if boolean == 1:
                boolean -= 1
                ax.text(point['x'], point['y']+0.002, str(point['val']))
            else:
                ax.text(point['x']-12, point['y'], str(point['val']))
        else:
            ax.text(point['x'], point['y']+0.002, str(point['val']))
```

```python
df_RO = df2[(df2.dataset == 'Deezer - Romania')]
plt.figure(figsize=(6,6))
label_point(df_RO.nb_clusters, df_RO.modularity, df_RO.algorithm, plt.gca())
sns.set()
sns.scatterplot(x='nb_clusters', y='modularity', size='time',
                hue='algorithm', sizes=(200, 1000), legend='brief',
                data=df_RO, palette = 'bright')
plt.legend([])
plt.show()
```
