```python
import pandas as pd
```

```python
df_dict = {'Dataset': ['artists', 'athletes', 'company', 'government',
                       'new_sites', 'politicians', 'public_figure', 'tv_show',
                       'croatia', 'hungary', 'romania'],
           'Edges': [819306, 86858, 52310, 89455,
                     206259, 41729, 67114, 17262,
                     498202, 222887, 125826],
           'Nodes': [50515, 13866, 14113, 7057,
                     27917, 5908, 11565, 3892,
                     54573, 47538, 41773]}
df = pd.DataFrame(df_dict).set_index('Dataset')
```

```python
df['Edges_%'] = df.Edges / df.Edges.max() * 100
df['Nodes_%'] = df.Nodes / df.Nodes.max() * 100
df['ETA (mins)'] = (df.Nodes ** 1.645) / (41773 ** 1.645) * 100
df = df[['Edges', 'Edges_%', 'Nodes', 'Nodes_%', 'ETA (mins)']]
df = df.astype('int')
df.sort_values('Nodes', ascending = False, inplace = True)
df
```
